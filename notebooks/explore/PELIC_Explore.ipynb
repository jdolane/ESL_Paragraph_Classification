# Explore the PELIC Dataset
This notebook is to explore the PELIC dataset, which has been compiled, cleaned, augmented, and balanced.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/PELIC_balanced.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.level.value_counts()
```

```python
df.L1.value_counts()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='level', hue='L1')
plt.title('Distribution of L1 by Level')
plt.xlabel('Level')
plt.ylabel('Count')
plt.legend(title='L1', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(7, 5))

# Plotting length by level
sns.boxplot(data=df, x='level', y='length')
plt.title('Length Distribution by Level')
plt.xlabel('Level')
plt.ylabel('Length')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))

# Plotting length by level
sns.boxplot(data=df, x='level', y='num_sentences')
plt.title('Number of sentences by Level')
plt.xlabel('Level')
plt.ylabel('Number of sentences')

plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='level', hue='is_augmented')
plt.title('Distribution of Augmented Data by Level')
plt.xlabel('Level')
plt.ylabel('Count')
plt.legend(title='Augmented Data', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
